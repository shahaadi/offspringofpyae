```python
import pandas as pd
import numpy as np
```

```python
ncs = pd.read_csv('ncs.csv', sep=';')
```

```python
VID_LEN = ncs.shape[0]
SONG_LEN = 240
labels = np.zeros((VID_LEN, SONG_LEN))
```

```python
start_s = ncs.start_s.values
end_s = ncs.end_s.values
```

```python
labels[np.arange(SONG_LEN)[np.newaxis, :] >= start_s[:, np.newaxis]] = 1
labels[np.arange(SONG_LEN)[np.newaxis, :] >= end_s[:, np.newaxis]] = 0
```

```python
idx = 1000

print(start_s[idx])
print(end_s[idx])

for i, label in enumerate(labels[idx]):
    print(i, labels[idx, i], sep='\t')
```

```python
import pickle
import blosc

pickled_data = pickle.dumps(labels)
compressed_pickle = blosc.compress(pickled_data)

with open(f'timestamp_labels/timestamp_labels.dat', 'wb') as f:
    f.write(compressed_pickle)
```
